```python
# Numerical arrays.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt
```

## Square Roots

$ a^2 = b \Rightarrow a = \sqrt{b}  $

```python
4**0.5
```

```python
# Test.
def issqrt(a, b):
    return np.isclose(a**2, b)
```

```python
issqrt(4**0.5, 4)
```

```python
fig, ax = plt.subplots()

x = np.linspace(-4.0, 4.0, 1000)

ax.plot(x, x**2, label='$y=x^2$')
#ax.plot(x, x**3, label='$y=x^3$')
ax.plot(x, np.abs(x), label='$y=|x|$')
ax.legend()
```

```python
def Cost(a, b):
    return (a**2 - b)**2
```

```python
for x in np.linspace(-4, 4, 9):
    print(f"x={x:6.2f}, Cost={Cost(x, 4):6.2f}")
```

```python
fig, ax = plt.subplots()

x = np.linspace(-6.0, 6.0, 1000)

ax.plot(x, Cost(x, 9), label='$Cost=(x^2-4)^2$')
ax.legend()
```

```python
# Amount to change x by.
dx = 0.01

# Initial guess.
x = 5.0

# Number of iterations.
niter = 1000

for i in range(niter):
    # Current cost.
    c = Cost(x, 11)

    # Cost if we increase x.
    c1 = Cost(x + dx, 11)

    # Cost if we decrease x.
    c2 = Cost(x - dx, 11)

    # Update x.
    if c1 < c:
        x += dx
    elif c2 < c:
        x -= dx
```
